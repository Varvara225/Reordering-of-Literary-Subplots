# Verbs

```python
!pip install pymorphy2
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
import os.path, pandas as pd
```

```python
# папка, куда помещались все обработанные файлы
folder = '/lemmas'

# папки с выгрузками из НКРЯ
dirs = os.listdir(folder)
```

```python
for d in dirs:
  frames = []
  verb_data = []
  for file in os.listdir(folder + '/' + d):
    df = pd.read_excel(folder + '/' + d + '/' + file)
    frames.append(df)
  data = pd.concat(frames)
  for v in pd.unique(data.Center):
    p = morph.parse(v.lower())[0]
    if p.tag.POS != 'ADJF':
      verb_data.append([p.normal_form, v.lower()])


  name = folder + '/' + 'lemmas' + '/' + 'lemmas_' + d + '.xlsx'
  df_lemmas = pd.DataFrame(verb_data, columns=['lemma', 'verb'])
  df_lemmas = df_lemmas.drop_duplicates(subset=['lemma'])
  df_lemmas.to_excel(name, index=False)

```

```python
# слова для различных конструкций

folder3 = '/toponyms'
space_constr = []
frames = []
for f in os.listdir(folder3):
  df = pd.read_excel(folder3 + '/' + f)
  frames.append(df)

total = pd.concat(frames[1:])
for n in pd.unique(total.Center):
  if type(n) != float:
    p = morph.parse(n.lower())[0]
    space_constr.append([p.normal_form, n.lower()])


name = 'space_constr.xlsx'
df_lemmas = pd.DataFrame(space_constr, columns=['lemma', 'noun'])
df_lemmas = df_lemmas.drop_duplicates(subset=['lemma'])
df_lemmas.to_excel(name, index=False)
```

```python
from ruwordnet import RuWordNet
wn = RuWordNet()
```

```python
verb_to_search = ['move', 'exist', 'appear', 'disappear', 'think', 'feel', 'locate', 'light', 'contact', 'behave', 'desire',
                  'smell', 'talk', 'impact', 'change', 'place']
```

```python
folder = '/lemmas'
```

```python
for v in verb_to_search:
  verb_data = []
  for w in wn.get_en_synsets(v):
    if '-v' in w.id:
      for s in w.ili:
        if s.title != 'МОТИВИРОВАТЬ' and s.title != 'ЗАДЕТЬ (ВЗВОЛНОВАТЬ)' and s.title != 'ОБЩЕНИЕ, СВЯЗЬ' and s.title != 'ОБМЕНЯТЬСЯ, ПРОИЗВЕСТИ ОБМЕН' and s.title != 'ОЦЕНИТЬ, ДАТЬ ОЦЕНКУ':
          # print(v, s.hyponyms)
          for hyp in s.hyponyms:
            for i in hyp.senses:
              verb_data.append(i.lemma.lower())
  name = folder + '/' + 'lemmas_ruwordnet_' + v + '.xlsx'
  df_lemmas = pd.DataFrame(verb_data, columns=['lemma'])
  df_lemmas = df_lemmas.drop_duplicates(subset=['lemma'])
  df_lemmas.to_excel(name, index=False)
```
